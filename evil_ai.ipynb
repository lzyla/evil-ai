```python
def get_posts(r):
    current_posts = []
    for section in r['sections']:
        posts = section['layout_content']['medias']

        for post in posts:
            post_datetime = post['media']['taken_at']
            post_datetime = datetime.datetime.fromtimestamp(post_datetime)
            print(post_datetime)
            #photo = post['media']['image_versions2']['candidates'][0]['url']

            if post_datetime > datetime.datetime(2024, 2, 3, 16, 30, 0) and post_datetime < datetime.datetime(2024, 2, 3, 16, 50, 0):
                current_posts.append(post)

            if post_datetime < datetime.datetime(2024, 2, 3, 16, 30, 0):
                print('Finished!')
                return current_posts, True
    return current_posts, False

posts_list = []

page_ix = 1
next_max_id = ''
next_page_id = ''
# Loop through all posts and download the photos between 1st of August 2023 until 31st of August 2023
while True:
    r = requests.post('https://www.instagram.com/api/v1/locations/%s/sections/' % location_id, headers=headers, data={
        'surface': 'grid',
        'tab': 'recent',
        'max_id': next_max_id,
        'page': str(page_ix)
    }).json()

    #next_page_id = r['next_page_id']
    page_ix = r['next_page']
    next_media_ids = r['next_media_ids']
    next_max_id = r['next_max_id']

    posts, status = get_posts(r)
    posts_list += posts
    if status:
        break
    page_ix += 1

# Save the list of posts as pickle file
import pickle
with open('posts_list.pkl', 'wb') as f:
    pickle.dump(posts_list, f)

# Download the photo from the url and save on disk in full resolution
photo_counter = 0
for post in posts_list:
    photo = post['media']['image_versions2']['candidates'][0]['url']
    r_photo = requests.get(photo, stream=True)
    with open('photos/%s.jpg' % photo_counter, 'wb') as f:
        time.sleep(2)
        f.write(r_photo.content)
        print('Downloaded %s' % photo)
    photo_counter += 1
```
